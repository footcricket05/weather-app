## Exploratory Data Analysis (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load cleaned data
data = pd.read_csv(r'C:\users\shaurya\Downloads\weather_forecasting_project\data\processed\cleaned_weather_data.csv')

# Print column names to verify
print(data.columns)
```

```python
# Parse the 'last_updated' column as a datetime object
if 'last_updated' in data.columns:
    data['last_updated'] = pd.to_datetime(data['last_updated'])
else:
    raise ValueError("The column 'last_updated' is missing from the dataset.")

# Line plot for temperature trends
plt.figure(figsize=(10, 5))
sns.lineplot(data=data, x='last_updated', y='temperature_celsius')  # Ensure 'temperature_celsius' exists in your dataset
plt.title('Temperature Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.show()
```
